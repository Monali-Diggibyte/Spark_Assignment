```
df_user = spark.read.format("csv").options(header= True, inferschema= True, sep="," ).load("/FileStore/tables/user.csv")

display(df_user)
```

```
df_transaction = spark.read.format("csv").options(header= True, inferschema= True, sep="," ).load("/FileStore/tables/transaction.csv")

display(df_transaction)
```

```
df_join = df_user.join(df_transaction, df_user.user_id == df_transaction.userid, "rightouter").orderBy(df_user.user_id)

display(df_join)

df_join.printSchema()
```

```
# print(df_join.countDistinct("location"))

print("Tota Count: " + str(df_join.distinct().count()))

df_join.select("location ").distinct().show()
print("Count of Unique Locations where each Product is sold: "+ str(df_join.select("location ").distinct().count()))
```

```
df2 = df_join.select('user_id', 'emailid', 'product_description').distinct().orderBy('user_id')

display(df2)
```

```
df3 = df_join.select('user_id', 'product_description', 'price').distinct().orderBy('user_id')

display(df3)
```
